```python
RP_IDS = [4, 20, 24]
```

```python
GROUP_TOP = "Top"
GROUP_BOTTOM = "Bottom"
GROUP_HORIZONTAL = "Horizontal"

GROUPS = [GROUP_TOP, GROUP_BOTTOM, GROUP_HORIZONTAL]

GROUP_STRING_TO_RP_ID_LAST_DIGITS = {
    GROUP_TOP : [0, 4],
    GROUP_BOTTOM : [1, 5],
    GROUP_HORIZONTAL : [2, 3]
}

GROUP_STRING_TO_GROUP_IDS = {
    GROUP_TOP : [1, 4],
    GROUP_BOTTOM : [2, 5],
    GROUP_HORIZONTAL : [3, 6]
}

# 0 4 TOP
# 1 5 BOT
# 2 3 HOR

def rp_group(rpID):
    if (rpID % 10) % 6 in GROUP_STRING_TO_RP_ID_LAST_DIGITS[GROUP_TOP]:
        return GROUP_TOP
    elif (rpID % 10) % 6 in GROUP_STRING_TO_RP_ID_LAST_DIGITS[GROUP_BOTTOM]:
        return GROUP_BOTTOM
    else:
        return GROUP_HORIZONTAL
```

```python
RP_ID_TO_BINS = {}
X_BINS = 'X_BINS' 
Y_BINS = 'Y_BINS' 
STEP = 3.0


for rpID in RP_IDS:
    if rp_group(rpID) == GROUP_TOP:
        x_bins = np.arange(-15.0, 15.0, STEP)
        y_bins = np.arange(0.0, 30.0, STEP)
    elif rp_group(rpID) == GROUP_BOTTOM:
        x_bins = np.arange(-15.0, 15.0, STEP)
        y_bins = np.arange(-30.0, 0.0, STEP)
    else:
        x_bins = np.arange(0.0, 30.0, STEP)
        y_bins = np.arange(-15.0, 15.0, STEP)
    
    RP_ID_TO_BINS[rpID] = {}
    RP_ID_TO_BINS[rpID][X_BINS] = x_bins
    RP_ID_TO_BINS[rpID][Y_BINS] = y_bins
```

### CUT OFF WRONG HITS 

```python
X_MIN = -20.0
X_MAX = 20.0
Y_MIN = 0.0
Y_MAX = 40.0
```

```python
def get_percent(a, b):
    return float(a) / b * 100


'''
Printing dataframe:
    rpID, track_type, all_hits, 
        frame_cut_no, frame_cut_pp, 
        cut_det_no, cut_det_pp,
        cut_avg_dis_no, cut_avg_dist_pp,
        cut_all_no, cut_all_pp
                                
W ten sposób potrzebuje 4 rekordy per rp.
'''

def log_cut_statistics(rpID, track_type, all_hits,
                       frame_cut_no, frame_cut_pp, 
                       cut_det_no, cut_det_pp,
                       cut_avg_dis_no, cut_avg_dist_pp,
                       cut_all_no, cut_all_pp):
    print('{} {} {} {} {} {} {} {} {} {} {}'.format(rpID, track_type, all_hits,
                       frame_cut_no, frame_cut_pp, 
                       cut_det_no, cut_det_pp,
                       cut_avg_dis_no, cut_avg_dist_pp,
                       cut_all_no, cut_all_pp))



def get_cut_df(df, rpID, cut_out_det_tracks=False, dist_avg_cut=None, track_type=None):
    print("Cutting rp {} hits".format(rpID))
    
    df_len = len(df)
    original_length = len(df)

    # CUT TRACKS OUT OF FRAME
    print("\tDf len: {}".format(df_len))
    cut_df = df.loc[(df['x'] > X_MIN) & 
                    (df['x'] < X_MAX) &
                    (df['y'] > Y_MIN) & 
                    (df['y'] < Y_MAX)]
    
    cut_len_1 = len(cut_df)
    cut_df_diff = df_len - cut_len_1
    print("\tCut to fit frame: {}, Percent: {}".format(cut_df_diff, get_percent(cut_df_diff, df_len)))


    # CUT OUT DET TRACKS
    if cut_out_det_tracks:
        cut_df = cut_df.loc[(cut_df['tracks_out_det_no'] == 0)]
        
    cut_len_2 = len(cut_df)
    cut_12_diff = cut_len_1 - cut_len_2
    print("\tCut by det in and out: {}, Percent: {}".format(cut_12_diff, get_percent(cut_12_diff, df_len)))
    
    
    # CUT TRACKS BY AVG DIST
    if dist_avg_cut is not None:
        cut_df = cut_df.loc[(cut_df['dist_avg'] < dist_avg_cut)]
    
    cut_len_3 = len(cut_df)
    cut_23_diff = cut_len_2 - cut_len_3
    print("\tCut by dist avg: {}, Percent: {}".format(cut_23_diff, get_percent(cut_23_diff, df_len)))
        
    cut_3df_diff = df_len - cut_len_3
    
    print("\tPlotted records: {}".format(cut_len_3))
    print("\tOverall cut records: {}. Percent: {}".format(cut_3df_diff, get_percent(cut_3df_diff, df_len)))
    log_cut_statistics(rpID, track_type, df_len,
                       cut_df_diff, get_percent(cut_df_diff, df_len),
                       cut_12_diff, get_percent(cut_12_diff, df_len),
                       cut_23_diff, get_percent(cut_23_diff, df_len),
                       cut_3df_diff, get_percent(cut_3df_diff, df_len))
    
    return cut_df
```
