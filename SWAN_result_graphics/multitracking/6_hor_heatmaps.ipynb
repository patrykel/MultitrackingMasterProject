```python
RP_ID_TO_DF = {
 103: RP_103_DF,
 122: RP_122_DF,
 123: RP_123_DF
}
```

```python
RP_IDS = [103, 122, 123]
```

```python
GROUP_TOP = "Top"
GROUP_BOTTOM = "Bottom"
GROUP_HORIZONTAL = "Horizontal"

GROUPS = [GROUP_TOP, GROUP_BOTTOM, GROUP_HORIZONTAL]

GROUP_STRING_TO_RP_ID_LAST_DIGITS = {
    GROUP_TOP : [0, 4],
    GROUP_BOTTOM : [1, 5],
    GROUP_HORIZONTAL : [2, 3]
}

GROUP_STRING_TO_GROUP_IDS = {
    GROUP_TOP : [1, 4],
    GROUP_BOTTOM : [2, 5],
    GROUP_HORIZONTAL : [3, 6]
}

# 0 4 TOP
# 1 5 BOT
# 2 3 HOR

def rp_group(rpID):
    if (rpID % 10) % 6 in GROUP_STRING_TO_RP_ID_LAST_DIGITS[GROUP_TOP]:
        return GROUP_TOP
    elif (rpID % 10) % 6 in GROUP_STRING_TO_RP_ID_LAST_DIGITS[GROUP_BOTTOM]:
        return GROUP_BOTTOM
    else:
        return GROUP_HORIZONTAL
```

```python
RP_ID_TO_BINS = {}
X_BINS = 'X_BINS' 
Y_BINS = 'Y_BINS' 
STEP = 3.0


for rpID in RP_IDS:
    if rp_group(rpID) == GROUP_TOP:
        x_bins = np.arange(-15.0, 15.0, STEP)
        y_bins = np.arange(0.0, 30.0, STEP)
    elif rp_group(rpID) == GROUP_BOTTOM:
        x_bins = np.arange(-15.0, 15.0, STEP)
        y_bins = np.arange(-30.0, 0.0, STEP)
    else:
        x_bins = np.arange(0.0, 30.0, STEP)
        y_bins = np.arange(-15.0, 15.0, STEP)
    
    RP_ID_TO_BINS[rpID] = {}
    RP_ID_TO_BINS[rpID][X_BINS] = x_bins
    RP_ID_TO_BINS[rpID][Y_BINS] = y_bins
```

### CUT OFF WRONG HITS 

```python
X_MIN = 0
X_MAX = 50
Y_MIN = -25
Y_MAX = 25
```

```python
def get_cut_df(df, rpID):
    print("Cutting rp {} hits".format(rpID))
    original_length = len(df)
    print("\tOriginal df lenghth: {}".format(original_length))
    cut_df = df.loc[(df['x'] > X_MIN) & 
                    (df['x'] < X_MAX) &
                    (df['y'] > Y_MIN) & 
                    (df['y'] < Y_MAX)]
    
    cut_df_length = len(cut_df)
    cut_rec_number = original_length - cut_df_length
    cut_rec_percent = float(cut_rec_number) / original_length * 100
    print("\tPlotted records: {}".format(cut_df_length))
    print("\tCut records: {}. Percent: {}".format(cut_rec_number, cut_rec_percent))
    return cut_df
```

# HEATMAPS 

### FUNCTIONS 

```python
_11_TYPE = '_11_1X1_3RP_1X1'
_12_TYPE = '_12_1X1_3RP_MORE'
_21_TYPE = '_21_2X2_3RP_ENOUGH'
```

```python
def filter_by_type(df=None, track_type=_11_TYPE):
    return df.loc[df['track_type'] == track_type]
```

```python
hits_png_base = '/eos/user/p/plawski/SWAN_projects/RzymskiKociolek/Histograms/analysis_multi_single/results/multi/hits_png/'

def plot_by_type(track_type=_11_TYPE):
    for rpID in RP_IDS:
        df = RP_ID_TO_DF[rpID]

        # CUT IT (FILTER OUT DISTANT TRASHES) AND FILTER BY TYPE
        if track_type is not None:
            df = filter_by_type(df=df, track_type=track_type)
        
        df = get_cut_df(df, rpID)
        

        # TODO FILTER OUT EXTREME CASES
        figure(figsize=(12, 9), dpi=100)

        # create data
        x = np.array(df['x'])
        y = np.array(df['y'])

        plot_title = 'Track hits in RP {} ({}).'.format(rpID, rp_group(rpID))
        
        if track_type is not None:
            plot_title = plot_title + ' Tracks type: {}'.format(track_type)
        else:
            plot_title = plot_title + ' Tracks type: ALL'
        
        plt.xlabel('x [mm]')
        plt.ylabel('y [mm]')
        plt.title(plot_title)
        # bins = (50,50)
        # x_bins = np.arange(-2.0, 2.0, 0.1)
        # y_bins = np.arange(-25.0, -4.5, 0.1) 
        # bins = [x_bins, y_bins]

        x_bins = RP_ID_TO_BINS[rpID][X_BINS]
        y_bins = RP_ID_TO_BINS[rpID][Y_BINS]

        
        plt.hist2d(x, y, bins = (400, 400), cmap=plt.cm.jet, norm=LogNorm())
        plt.colorbar()
        
        # Create filename
        hits_png_filename = hits_png_base + plot_title.replace(' ', '_')
#         if track_type is not None:
#             hits_png_filename = hits_png_filename + '_' + track_type
        hits_png_filename = hits_png_filename + '.png'
        
        plt.savefig(hits_png_filename)
        
        plt.show()
```
